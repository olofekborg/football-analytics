```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
```

```python
df = pd.read_csv('2013-2023.csv')
```

```python
def set_color(team):
    if team  == 'IFK Göteborg':
        return '#234B9A'
    elif team == 'AIK':
        return '#08090e'
    elif team == 'Hammarby':
        return '#008245'
    elif team == 'Djurgården':
        return '#ef0000' # Red
        #return '#0d1526' # Blue
    elif team == 'Malmö FF':
        return '#0098e0'
    elif team == 'BK Häcken':
        return '#f5d400'
    elif team == 'BP':
        return '#f1000a'
    elif team == 'Sirius':
        return '#005cb3'
    elif team == 'Halmstads BK':
        return '#0057a8'
    elif team == 'Elfsborg':
        return '#fadb04'
    elif team == 'Norrköping':
        return '#0052a5'
    elif team == 'Degerfors IF':
        return '#fc0404'
    elif team == 'Mjällby':
        return '#ffd400'
    elif team == 'Värnamo':
        return '#005aa4'
    elif team == 'Varberg':
        return '#00943b'
    elif team == 'Kalmar':
        return '#e3002d'
    
    
def set_markersize(team):
    if team  == 'IFK Göteborg':
        return 8
    else:
        return 5

def set_alpha(team):
    if team  == 'IFK Göteborg':
        return 1
    else:
        return 0.6
    
    
def set_linestyle(team):
    if team  == 'IFK Göteborg':
        return '--'
    else:
        return ':'
    
    
def set_linewidth(team):
    if team  == 'IFK Göteborg':
        return 3
    else:
        return 2
    

def set_zorder(team):
    if team  == 'IFK Göteborg':
        return 3
    else:
        return 2
```

## Year by year - Big teams

```python
fig, ax = plt.subplots(figsize=(18,12))

for team_name in df.team.unique():
    df_plot = df.loc[df.team == team_name]

    plt.plot(df_plot.year, df_plot.attendance, 
         color=set_color(team_name),
             linewidth=set_linewidth(team_name),
             linestyle=set_linestyle(team_name),
             marker='o',
             markersize=set_markersize(team_name),
             alpha=set_alpha(team_name),
             label=team_name
            )
    
    if team_name == 'IFK Göteborg':
        for _, row in df_plot.iterrows():
            v_dist = -1500 if row.year == 2023 else 700
            h_dist = -0.25 if row.year == 2023 else 0.1
            ax.annotate(row.attendance, (row.year + h_dist, row.attendance + v_dist),va='top',size=14, zorder=4,
                        bbox=dict(boxstyle='round', fc='#f2f2f2', ec='none', alpha=0.7))

    #url = logo_url[team_name]
    #response = requests.get(url)
    img = club_icon = Image.open(urllib.request.urlopen(logo_url[team_name]))
    im = OffsetImage(img, zoom=1)
    
    x = df_plot.year.values[-1] if team_name != 'Djurgården' else df_plot.year.values[-1] + 0.25 
    y = df_plot.attendance.values[-1]
    ab = AnnotationBbox(im, (x, y), xycoords='data', frameon=False)
    ax.add_artist(ab)

ax.xaxis.set_ticks(np.arange(2013, 2024))
ax.spines[['top', 'right']].set_visible(False)
ax.grid(axis='y')
ax.set_ylim([0, 32000])
#ax.legend(bbox_to_anchor=(1, 1.05),  prop={'size': 16})

ax2 = fig.add_axes([0.11,0.9,0.1,0.1]) # badge
ax2.axis("off")
url = 'https://anglarna.se/content/files/50a__rdelar.png'
response = requests.get(url)
img = Image.open(BytesIO(response.content))
ax2.imshow(img)

fig.text(0.2,0.92,'Bortastöd\n', fontsize=32, fontweight="bold")
fig.text(0.1955,0.92,' Allsvenskan 2013-2023', fontsize=24, fontweight="regular")

plt.savefig('bortastöd.png', bbox_inches='tight')
plt.show()
```

## 2023 - IFK Göteborg

```python
df_ifk = pd.read_csv('2023_ifk.csv')
```

```python
df_ifk['color'] = df_ifk.opponent.apply(lambda x: set_color(x))
```

```python
def format_function(x):
    return '{:.0f}\n{:.1f}%'.format(total*x/100 , x)


total = df_ifk.attendance.sum()

fig, ax = plt.subplots(figsize=(12,12))

_, texts, _ = ax.pie(df_ifk.attendance, labels=df_ifk.opponent, autopct=format_function, pctdistance=0.8, labeldistance=1.15, startangle=90,counterclock=False,
       textprops={'size': 16, 'color':'white'}, colors=df_ifk.color)
#'bbox': dict(boxstyle='round4',fc='white', ec='none', alpha=0.4)
for item in texts:
    xy = (item.get_position()[0],item.get_position()[1])

    img = club_icon = Image.open(urllib.request.urlopen(logo_url[item.get_text()]))
    im = OffsetImage(img, zoom=1)

    ab = AnnotationBbox(im, xy, xycoords='data', frameon=False, boxcoords='offset points', pad=0)
    ax.add_artist(ab)
    
    item.remove()
    
plt.savefig('ifk_2023_cirkel.png', bbox_inches='tight')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(18,12))

plt.bar(df_ifk.opponent, df_ifk.attendance, 
        color=df_ifk.color,
        #color='#234B9A'
)

for _, row in df_ifk.iterrows():
    ax.annotate(row.attendance, (row.opponent, row.attendance + 60),va='top',ha='center',size=14, zorder=4,
                bbox=dict(boxstyle='round', fc='#f2f2f2', ec='none', alpha=0.7))
    
    img = club_icon = Image.open(urllib.request.urlopen(logo_url[row.opponent]))
    im = OffsetImage(img, zoom=1)

    x = row.opponent

    ab = AnnotationBbox(im, (x, 0), xybox=(0., -35.), xycoords='data', frameon=False, boxcoords="offset points", pad=0)
    ax.add_artist(ab)

ax.spines[['top', 'right']].set_visible(False)
ax.grid(axis='y')
ax.set_xticklabels([])

ax2 = fig.add_axes([0.11,0.9,0.1,0.1]) # badge
ax2.axis("off")
url = 'https://anglarna.se/content/files/50a__rdelar.png'
response = requests.get(url)
img = Image.open(BytesIO(response.content))
ax2.imshow(img)

fig.text(0.2,0.92,'Bortastöd IFK Göteborg\n', fontsize=32, fontweight="bold")
fig.text(0.1955,0.92,' Allsvenskan 2023', fontsize=24, fontweight="regular")

plt.savefig('ifk_2023_stapel_färg.png', bbox_inches='tight')
plt.show()
```
